#### Kiểm tra thứ tự thời gian
Thứ tự đúng phải là: purchase ≤ approved ≤ carrier ≤ customer.
Có những đơn KHÔNG null nhưng giá trị vô lý (vd: giao cho carrier trước khi
được duyệt). Đây là lỗi logic, không phải lỗi thiếu dữ liệu -> cần loại.

```python
# Chỉ xet đơn delivered đã có đủ mốc
d_mask = orders['order_status']=='delivered'

invalid = orders[
    d_mask & (
        (orders['order_delivered_carrier_date'] < orders['order_approved_at']) |
        (orders['order_delivered_customer_date'] < orders['order_delivered_carrier_date'])
    )
]
print('Số đơn có mốc phi logic:', len(invalid))
```

```python
orders = orders[~(
    (orders['order_status']=='delivered') & (
        (orders['order_delivered_carrier_date'] < orders['order_approved_at']) |
        (orders['order_delivered_customer_date'] < orders['order_delivered_carrier_date'])
    )
)]
```

```python
geolocation.shape
```

```python
geolocation['geolocation_zip_code_prefix'].nunique()
```

```python
geolocation['geolocation_zip_code_prefix'] = (geolocation['geolocation_zip_code_prefix'].astype(str).str.zfill(5))
```

```python
for col in ['geolocation_city', 'geolocation_state']:
    geolocation[col] = (geolocation[col].astype(str).str.strip())
```

```python
geo_clean = (geolocation
    .groupby('geolocation_zip_code_prefix')
    .agg(geo_lat=('geolocation_lat','median'),
        geo_lng=('geolocation_lng','median'),
        geo_state=('geolocation_state', lambda s: s.mode().iloc[0]))
    .reset_index()
    .rename(columns={'geolocation_zip_code_prefix':'zip_prefix'}))
```
